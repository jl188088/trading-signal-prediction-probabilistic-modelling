```python
#Libraries

import pandas as pd
import matplotlib.pyplot as plt
from hmmlearn.hmm import GaussianHMM
```

```python
# Load data
df = pd.read_csv("../data/raw/data.csv")
```

```python
# Compute returns
df["returns"] = df["Close"].pct_change()
df.dropna(inplace=True)
```

```python
# Train HMM
model = GaussianHMM(n_components=3, n_iter=1000)
X = df["returns"].values.reshape(-1, 1)
model.fit(X)
```

```python
# Predict states
states = model.predict(X)
probs = model.predict_proba(X)

df["state"] = states
df["prob_bull"] = probs[:, 1]

# Plot states
plt.figure()
plt.scatter(range(len(df)), df["Close"], c=df["state"])
plt.title("Market Regimes (HMM)")
plt.show()
```

```python
# Plot probabilities
plt.figure()
plt.plot(df["prob_bull"])
plt.title("Bull Probability")
plt.show()
```
